```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
```

#### import dataset

```python
dataset=pd.read_excel('dataset.xlsx')
```

#### EDA process

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['review'] = (
    dataset['title'].fillna('').astype(str) + '  :  ' +
    dataset['body'].fillna('').astype(str)
)
dataset = dataset.drop([ 'body'], axis=1)
```

```python
# REMOVE EMPTY
dataset = dataset[dataset['review'].str.strip() != ""]

```

```python
# REMOVE NON-ENGLISH (Marathi etc.)
dataset = dataset[dataset['review'].str.contains(r'[a-zA-Z]', regex=True)]
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.drop_duplicates(inplace=True)
dataset.shape
```

```python
for col in dataset.columns:
    print('=============================')
    print(f"Column: {col}")

    print("Data Type:", dataset[col].dtype)
    print("Missing Values:", dataset[col].isnull().sum())
    print("Unique Values:", dataset[col].nunique())

    print("Top Values:")
    print(dataset[col].value_counts().head(5))

    print('=============================')
```

```python
def get_sentiment(rating):
    if rating>=4:
        return 'POSITIVE'
    elif rating==3:
        return 'NEUTRAL'
    else:
        return 'NAGATIVE'
```

```python
dataset['Sentiment']=dataset['rating'].apply(get_sentiment)
```

```python
print('sentiment Distribution:-')
print(dataset['Sentiment'].value_counts())
```

#### frequency distribution between sentiments

```python
plt.figure(figsize=(8,7))
sns.barplot( dataset['Sentiment'].value_counts())
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
dataset['review_length']=dataset['review'].apply(lambda x:len(str(x)))
```

```python
dataset.head()
```

```python
print("\nReview Length Stats:\n")
print(dataset['review_length'].describe())
```

#### find distribution ,skewness between datset

```python
num_cols = dataset.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure(figsize=(8,7))
    sns.displot(dataset[col], bins=50, kde=True)
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(alpha=0.3)
    plt.show()
```

#### outlier handling

```python
plt.figure(figsize=(8,7))
sns.boxplot(data=dataset[num_cols])
plt.title('outlier detection')
plt.show()
```

```python
Q1 = dataset[num_cols].quantile(0.25)
Q3 = dataset[num_cols].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
for col in num_cols:
    lower = lower_bound[col]
    upper = upper_bound[col]

    dataset[col] = dataset[col].clip(lower, upper)
```

```python
plt.figure(figsize=(8,7))
sns.boxplot(data=dataset[num_cols])
plt.title('outlier detection')
plt.show()
```

```python
sns.countplot(data=dataset, x='rating')
plt.title('Count plot for rating')
plt.show()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python

nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
from wordcloud import WordCloud
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_review(text):
    text = str(text).lower()
   # REMOVE URL
    text = re.sub(r'http\S+', '', text)

    # REMOVE HTML
    text = re.sub(r'<.*?>', '', text)

    # REMOVE SPECIAL CHARACTERS
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    words = [lemmatizer.lemmatize(w) for w in words]
    return words   # return tokens for EDA
```

```python
# tokenization
dataset['tokens'] = dataset['review'].apply(clean_review)
```

```python
# clean_text
dataset['clean_review']=dataset['tokens'].apply(lambda x:" ".join(x))
```

```python
# WORD FREQUENCY
all_words = []
for t in dataset['tokens']:
    all_words.extend(t)

freq = Counter(all_words)

print("Top Words:", freq.most_common(10))
```

```python
# PLOT WORDS
top_words = freq.most_common(10)
words = [i[0] for i in top_words]
counts = [i[1] for i in top_words]

plt.figure(figsize=(8,7))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top Words")
plt.show()
```

```python
#  word cloud
text=" ".join(dataset['clean_review'])
wordcloud=WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(text)

plt.figure(figsize=(8,7))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("word colud")
plt.show()
```

```python
# REVIEW LENGTH
dataset['review_length'] = dataset['review'].apply(len)

sns.boxplot(x='Sentiment', y='review_length', data=dataset)
plt.title("Review Length vs Sentiment")
plt.show()
```

```python
Q1 = dataset['review_length'].quantile(0.25)
Q3 = dataset['review_length'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
dataset['review_length'] = dataset['review_length'].clip(lower_bound, upper_bound)
```

```python
sns.boxplot(x='Sentiment', y='review_length', data=dataset)
plt.title("Review Length vs Sentiment (After Outlier Treatment)")
plt.show()
```

# MODEL BULDING

```python
import pandas as pd
import re
import pickle
import nltk
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

!pip install gensim
from gensim.models import Word2Vec
```

```python
dataset=pd.read_excel('clean_data.xlsx')
dataset.head()
```

```python
stop_word=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def clean_review(text):
    text=str(text).lower()

       # REMOVE URL
    text = re.sub(r'http\S+', '', text)

    # REMOVE HTML
    text = re.sub(r'<.*?>', '', text)

    # REMOVE SPECIAL CHARACTERS
    text = re.sub(r'[^a-zA-Z]', ' ', text)

    words=text.split()
    words=[w for w in words if w not in  stop_word]
    words=[lemmatizer.lemmatize(w) for w in words ]

    return " ".join(words)
```

```python
# =========================
# CLEAN REVIEW
# =========================
dataset['clean_review']=dataset['review'].apply(clean_review)
```

```python
# =========================
# WORD2VEC
# =========================
tokens=dataset['clean_review'].apply(lambda x:x.split())
```

```python
word2vec_model=Word2Vec(
    sentences=tokens,
    vector_size=100,
    window=5,
    min_count=2
)
```

```python
print(word2vec_model.wv.most_similar('good'))
```

```python
# =========================
# TF-IDF
# =========================
vectorizer=TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2),
    min_df=5,
    max_df=0.85,
    sublinear_tf=True
)
x=vectorizer.fit_transform(dataset['clean_review'])
y=dataset['Sentiment']
```

```python
# split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# model
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
# prediction

y_pred=model.predict(x_test)
```

```python
# evaluation
print("Train Accuracy:")
print(model.score(x_train, y_train)*100)

print("Test Accuracy:")
print(model.score(x_test, y_test)*100)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# =========================
# SAVE MODEL
# =========================
pickle.dump(
    model,
    open('model.pkl','wb')
)
```

```python
pickle.dump(
    vectorizer,
    open('vectorizer.pkl','wb')
)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("=== Step 1: Basic Metrics ===")
print(f"Accuracy  : {accuracy_score(y_test, y_pred):.3f}")
print(f"Precision : {precision_score(y_test, y_pred, average='weighted'):.3f}")
print(f"Recall    : {recall_score(y_test, y_pred, average='weighted'):.3f}")
print(f"F1-Score  : {f1_score(y_test, y_pred, average='weighted'):.3f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print("\n=== Step 3: Confusion Matrix ===")
cm = confusion_matrix(y_test, y_pred)
print(cm)
# [[TN, FP]
#  [FN, TP]]

# Plot
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Pred Neg', 'Pred Pos'],
            yticklabels=['Actual Neg', 'Actual Pos'])
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import LabelBinarizer # Import LabelBinarizer

print("\n=== Step 4: ROC-AUC ===")

# Calculate predicted probabilities
y_pred_proba = model.predict_proba(x_test)

# Calculate ROC-AUC for multiclass
# 'ovr' (One-vs-Rest) strategy and 'weighted' average are suitable for multiclass imbalanced datasets
roc_auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr', average='weighted')
print(f"ROC-AUC (weighted, ovr): {roc_auc:.3f}")  # 1.0=Perfect, 0.5=Useless

# Plot ROC Curves for each class (One-vs-Rest)
plt.figure(figsize=(10, 8))
lb = LabelBinarizer()
y_test_binarized = lb.fit_transform(y_test)

# Ensure class labels are in the same order as model.classes_
class_labels = model.classes_

for i, class_label in enumerate(class_labels):
    fpr, tpr, _ = roc_curve(y_test_binarized[:, i], y_pred_proba[:, i])
    roc_auc_class = roc_auc_score(y_test_binarized[:, i], y_pred_proba[:, i])
    plt.plot(fpr, tpr, label=f'ROC curve for {class_label} (AUC = {roc_auc_class:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier (AUC = 0.50)')  # Random line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - One-vs-Rest for each class')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

print("\n=== Step 5: Error Analysis ===")

# Get the clean review texts corresponding to the test set
x_test_text = dataset.loc[y_test.index, 'clean_review'].reset_index(drop=True)

# Get the confidence of the predicted class
proba_confidence = y_pred_proba.max(axis=1)

df_results = pd.DataFrame({
    'text': x_test_text,
    'actual': y_test.reset_index(drop=True),
    'pred': y_pred,
    'proba': proba_confidence
})

# False Positives for NEUTRAL (Actual: NAGATIVE, Predicted: NEUTRAL)
# Based on model.classes_ order: 'NAGATIVE' (index 0), 'NEUTRAL' (index 1), 'POSITIVE' (index 2)
fp = df_results[(df_results['actual']=='NAGATIVE') & (df_results['pred']=='NEUTRAL')]
print("Top 5 False Positives (Actual NAGATIVE, Predicted NEUTRAL):")
print(fp.sort_values('proba', ascending=False)[['text','proba']].head())

# False Negatives for NEUTRAL (Actual: NEUTRAL, Predicted: NAGATIVE)
fn = df_results[(df_results['actual']=='NEUTRAL') & (df_results['pred']=='NAGATIVE')]
print("\nTop 5 False Negatives (Actual NEUTRAL, Predicted NAGATIVE):")
print(fn.sort_values('proba', ascending=False)[['text','proba']].head())
```

```python
from sklearn.metrics import balanced_accuracy_score, matthews_corrcoef
from sklearn.metrics import precision_recall_curve
from sklearn.preprocessing import LabelBinarizer # Import LabelBinarizer if not already present

print("\n=== Step 6: Imbalanced Data Metrics ===")
print("Balanced Accuracy:", balanced_accuracy_score(y_test, y_pred))
print("MCC:", matthews_corrcoef(y_test, y_pred))  # -1 to +1. 0=random

# PR Curve for Multiclass (One-vs-Rest)
plt.figure(figsize=(10, 8))
# Reuse LabelBinarizer and binarized y_test from previous ROC-AUC cell if available
# If not available, initialize here: lb = LabelBinarizer(); y_test_binarized = lb.fit_transform(y_test)
# Ensure class_labels are consistent

for i, class_label in enumerate(model.classes_):
    precision, recall, _ = precision_recall_curve(y_test_binarized[:, i], y_pred_proba[:, i])
    plt.plot(recall, precision, label=f'PR curve for {class_label}')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve - One-vs-Rest for each class')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def evaluate_sentiment(y_test, y_pred, y_pred_proba, model_classes):
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=model_classes))

    # ROC-AUC for multiclass
    roc_auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr', average='weighted')
    print("ROC-AUC (weighted, ovr):", roc_auc)

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=model_classes,
                yticklabels=model_classes)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Use it
evaluate_sentiment(y_test, y_pred, y_pred_proba, model.classes_)
```
